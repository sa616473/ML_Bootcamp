```python
X = df.drop('benign_0__mal_1', axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
#Creating model and preventing overfitting
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense,Dropout
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
len(df.columns)-1
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))

#Since it is Binary Classification
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test,y_test))
```

```python
loss = pd.DataFrame(model.history.history)
loss.plot()
#This is overfitting
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))

#Since it is Binary Classification
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Using the callback will help us to detect overfitting 
# and stop the training early
early_stop = EarlyStopping(monitor = 'val_loss', mode='min', verbose=1, patience=25)
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test,y_test), callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
#We can use dropout layers as well to
# Randomly turn off some neurons to 
# stop overfitting
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5))


#Since it is Binary Classification
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, batch_size=128, validation_data=(X_test,y_test), callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
