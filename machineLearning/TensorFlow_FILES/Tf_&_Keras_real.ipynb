```python
#Feature enginerring process
```

```python
df = df.drop('id', axis=1)
```

```python
#Converting the date string to a date time object
df['date'] = pd.to_datetime(df['date'])
```

```python
df['date'].head()
```

```python
#Extracting the year
df['year'] = df['date'].apply(lambda date: date.year)
```

```python
df['month'] = df['date'].apply(lambda date: date.month)
```

```python
df.head()
```

```python
# lets see if there a price difference according to the month or year
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='month', y='price', data=df)
```

```python
df.groupby('year').mean()['price'].plot()
```

```python
df = df.drop('date', axis=1)
```

```python
df = df.drop('zipcode', axis=1)
```

```python
df.head()
```

```python
# When we see the year renoveted colunme we can see that
# as the year increases the value might be higher
df['yr_renovated'] = df['yr_renovated'].apply(lambda year: 1 if year > 0 else 0 )
```

```python
df['sqft_basement'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('price', axis=1).values #.values will return the numpy array format
```

```python
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
```

```python
#We usally scale our neurons according to 
# the incoming features
X_train.shape
```

```python
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))


model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')#adam optimizer
```

```python
model.fit(x=X_train, y=y_train, 
          validation_data= (X_test, y_test),
         batch_size=128, # smaller batch set => !overfitting
          epochs =400,
         )
```

```python
#Model evaluation and predictions
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
#model evaluation
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
print(mean_squared_error(y_test, predictions)**0.5)
print(mean_absolute_error(y_test, predictions))
print(explained_variance_score(y_test, predictions))
```

```python
plt.figure(figsize=(12,6))
plt.scatter(y_test, predictions)
plt.plot(y_test, y_test, 'r')
```
